A line is composed by an base point (origin) and a direction vector.

We can represent a line as a function.

Given the base point $\vec{b} = (b_x, b_y)$ and the direction vector $\vec{d} = (d_x, d_y)$.

$$
\vec{r}(t) = \vec{b} + t \vec{d}
$$

The (t) parameter is the distance from the base point to the direction vector ($\vec{d}$).

If t is 0, the point is the base point.

If t is 1, the point is the base point plus the direction vector.

If t is 2, the point is the base point plus the direction vector twice.

So t is a percentage of the direction vector.

## Intersection possibilities

If two lines are parallel, they will whether:

* Never intersect
* Intersect infinitely times

If two lines are not parallel, they will intersect at one point.

## Line representation

A line can be represented by:

$$
Ax + By = C
$$

Where the C is the distance from the origin to the line.

## Normal vector

The normal vector is $(A, B)$. Two scalar numbers.

If the normal vector of two lines are the same, the lines are parallel.

If we multiply the normal vector by a scalar, the line will be scaled.

## The C

The C is the distance from the origin to the line.

If we multiply the C by a scalar, the line will be translated, which means that it will be far away or closer to the origin.

> Changing C does not change the direction of the line, just its distance from the origin.

If two lines are parallel but C is different, they will never intersect.

If two lines are parallel and C is the same, they will intersect infinitely times.
